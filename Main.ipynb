```python
from Bio import SeqIO

# hanya ambil sequence Tahap 1
seq = SeqIO.read('sequence.fasta', 'fasta')
#print(seq.seq)
seq1=seq.seq

gb = SeqIO.read('sequence.gb','genbank')
seq_gb = gb.seq
seq2=gb.seq
print(seq_gb)
```

```python
# membuat dot plot kata dimas engga keluar

def dotplot(seqA , seqB):
    blank = ' '
    fill =chr(0x25A0)
    print(' |'+seqA)
    for i in seqB:
        print(i+'|', end = '')
        for j in seqA:
            if i == j:
                print(fill,end='')
            else :
                print(blank,end='')
        print()
dotplot(seqA, seqB)
```
